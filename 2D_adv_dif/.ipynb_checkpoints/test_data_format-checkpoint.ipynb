# Advection diffussion Equation:

```python
import numpy as np
import matplotlib.pyplot as plt

# Function for exact solution
def exact_solution_2D_advdiff(Re, t, L=2.0, Nx=128, Ny=128, c=1.0):
    nu = 1.0 / Re
    x_vals = np.linspace(-L, L, Nx)
    y_vals = np.linspace(-L, L, Ny)
    X, Y = np.meshgrid(x_vals, y_vals)
    if t <= 0:
        return X, Y, np.zeros_like(X)
    rx = X - c * t
    ry = Y
    r2 = rx**2 + ry**2
    denominator = 4.0 * nu * t
    amplitude = 1.0 / (4.0 * np.pi * nu * t)
    U = amplitude * np.exp(-r2 / denominator)
    return X, Y, U

# Function to generate the grid of plots
def generate_plots(re_range, t_range):
    # Divide re_range and t_range into 5 equal parts
    re_values = np.linspace(re_range[0], re_range[1], 10)
    t_values = np.linspace(t_range[0], t_range[1], 5)

    fig, axes = plt.subplots(10, 5, figsize=(20, 30))
    fig.subplots_adjust(hspace=0.4, wspace=0.4)

    for i, Re in enumerate(re_values):
        for j, t in enumerate(t_values):
            ax = axes[i, j]
            X, Y, U = exact_solution_2D_advdiff(Re, t, L=4.0, Nx=128, Ny=128, c=1.0)
            cf = ax.contourf(X, Y, U, levels=50, cmap='jet')
            ax.set_title(f"Re={Re:.2e}, t={t:.2f}")
            ax.set_xlabel('x')
            ax.set_ylabel('y')

    # Add colorbar to the last axis for reference
    cbar = fig.colorbar(cf, ax=axes[:, -1], location='right', shrink=0.7, label='u(x,y,t)')
    plt.show()

# Define ranges for Re and t
re_range = [0.01, 10]  # Range for Re (logarithmic scale)
t_range = [0.01, 2.0]  # Range for t

# Generate the plots
generate_plots(re_range, t_range)
```
